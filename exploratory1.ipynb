```python
import pymrio
```

```python
import country_converter as coco
```

```python
eoras = [
    pymrio.parse_eora26(year=year, path="data/" + str(year)).calc_all()
    for year in range(2015, 2018)
]
```

```python
for eora in eoras:
    eora.aggregate(
        region_agg=coco.agg_conc(
            original_countries="Eora",
            aggregates=["EU", {"Chile": "LE", "Bolivia": "LE", "Argentinia": "LE"}],
            missing_countries="NON_OECD",
        )
    )
```
